На основе выбранных продуктов из первого вебинара, выберите 2 A/B-теста и составьте план эксперимента, чтобы избежать Peeking Problem

Распишите, какие метрики и как вы будете анализировать

### 1.Гипотеза

Если делать рассылку пользователям, которые бросили корзины, с напоминанием, что в корзине лежат товары, это увеличит продажи на 15%, потому что некоторые люди могут отвлечься на другие хлопоты и забыть, что хотели оформить заказ.

### 2.Что делаем 

Контрольная версия: Не делаем рассылку пользователям бросившим корзину
Тестовая версия: Рассылаем напоминания пользователям о брошенных корзинах 

### 3.На каких пользователях тестируем
Тестируем на всех ользователях бросивших корзины (50% - не делаем рассылку, 50% делаем рассылку)

### 4.Ключевые метрики для оценки эксперимента
Процент конверсий в покупку (%CR) — primary. Процент показателя отказов, процент перехода в корзину.

##### Добавочные метрики: посещаемость, время на сайте, количество отписавшихся от рассылки

### 5.Выводы (Какие результаты получили, какие выводы сделали, какие решения приняли).
Если наш эксперимент будет положительным и мы зафиксируем ожидаемое улучшениев ключевых метриках и не посадим добавочные, то масштабируем изменение и «выкатываем» его на всех пользователей.
Если основные метрики падают или растут недостаточно либо добавочные метрики падают, откатываем эксперимент.
